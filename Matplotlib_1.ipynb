```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

%matplotlib inline
```

```python
years_of_exp=[1,2,3,4,5,6,7,8]
salary=[1000,2000,3000,4700,5000,6000,7000,8000]

```

```python
plt.figure()
plt.plot(years_of_exp,salary)
plt.show()
```

Line Plot

```python
plt.figure(figsize=(10,8),dpi=100)
plt.plot(years_of_exp,salary)
plt.show()
```

Scatter Plot


```python
plt.figure(figsize=(10,8),dpi=100)
plt.scatter(years_of_exp,salary,c="#ff00ff", s=40)
plt.show()
```

```python
plt.figure(figsize=(10,8),dpi=100)
plt.plot(years_of_exp,salary)
plt.title("Salary and Year of Experince Plot")
plt.xlabel("Experince in Job")
plt.ylabel("Salary")
plt.show()
```

```python
yearsof_exp=[1,2,3,4,5]
salary1=[10000,20000,34000,56798,45276]
salary2=[23867,30000,42000,52000,16000]
```

```python
plt.figure(figsize=(12,8),dpi=100)
plt.plot(yearsof_exp,salary1,'r-o',label="KPMG")
plt.plot(yearsof_exp,salary2,'b--o',label="KPM2G")
plt.title("Salaries of two organizations")
plt.xlabel("Salary of Organization1")
plt.ylabel("Years of exp")
plt.xticks(yearsof_exp)
plt.yticks(range(0,80000,10000))
plt.legend()
plt.show()
```

Bar Plot


```python
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates', 'Elderberries']
quantities = [120, 80, 90, 150, 100]

plt.figure(figsize=(12,5), dpi=100)
plt.bar(fruits, quantities, color='purple', width=0.4)
plt.yticks(range(0, 201, 20))
plt.show()
```

Pie Chart

```python
nums=[10,30,20,40,50]
plt.figure(figsize=(10,4),dpi=100)
plt.pie(nums)
plt.show()
```

```python
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates', 'Elderberries']
quantities = [120, 80, 90, 150, 100]
plt.figure(figsize=(10,6),dpi=100)
plt.pie(nums,labels=fruits,autopct='%1.2f%%')
plt.show()
```

```python
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates', 'Elderberries']
quantities = [120, 80, 90, 150, 100]
plt.figure(figsize=(10,6),dpi=100)
plt.pie(nums,labels=fruits,autopct='%1.2f%%',explode=[0,0.1,0,0.3,0])
plt.show()
```

Box Plot

```python
salaries = [30000, 35000, 40000, 45000, 50000, 55000,25000,50000,40000000, 60000, 65000, 70000, 120000]

plt.figure(figsize=(12,6), dpi=100)
plt.boxplot(salaries)
plt.show()

print('Q1 : ', np.quantile(salaries, 0.25))
print('Median (Q2) : ', np.quantile(salaries, 0.50))
print('Q3 : ', np.quantile(salaries, 0.75))
```

```python
import seaborn as sns
import pandas as pd
# Set Seaborn style settings for the plots
sns.set(rc={
    'figure.facecolor': 'white',    # Set the figure background color to white
    'axes.grid': True,              # Enable grid lines on the plot
    'axes.facecolor': '#95e4ed',    # Set the axes background color
    'axes.labelsize': 14,           # Set the font size for axis labels
    'axes.labelcolor': 'black',     # Set the color for axis labels
    'axes.titlesize': 16,           # Set the font size for the plot title
    'axes.titlecolor': 'black',     # Set the color for the plot title
    'xtick.labelsize': 10,          # Set the font size for x-axis tick labels
    'xtick.labelcolor': 'black',    # Set the color for x-axis tick labels
    'ytick.labelsize': 10,          # Set the font size for y-axis tick labels
    'ytick.labelcolor': 'black'     # Set the color for y-axis tick labels
})

# Create some sample data for the plot
np.random.seed(0)  # Set the seed for reproducibility of random numbers
data = pd.DataFrame({
    'x': np.linspace(0, 10, 100),  # Generate 100 values evenly spaced between 0 and 10 for x-axis
    'y': np.sin(np.linspace(0, 10, 100)) + np.random.normal(scale=0.5, size=100)  # Generate y values as sine of x with added noise
})

# Create a Seaborn line plot using the sample data
plt.figure(figsize=(10, 6))  # Set the size of the figure
sns.lineplot(data=data, x='x', y='y')  # Plot the line chart with 'x' as x-axis and 'y' as y-axis

# Customize the title and axis labels
plt.title('Sample Line Plot')  # Set the plot title
plt.xlabel('X Axis')  # Label for the x-axis
plt.ylabel('Y Axis')  # Label for the y-axis

# Display the plot
plt.show()  # Render and show the plot
```
